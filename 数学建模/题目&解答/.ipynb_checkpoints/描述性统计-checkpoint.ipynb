# 各类蔬菜月销售量变化

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
plt.rcParams["font.sans-serif"] = ["SimHei"]  # 设置字体
plt.rcParams["axes.unicode_minus"] = False  # 正常显示负号

df = pd.read_excel("蔬菜品类日销量.xlsx",index_col=0) # 蔬菜品类日销量.xlsx
```

```python
monthly_sum = df.resample('M').sum()
monthly_sum['Total'] = monthly_sum.iloc[:, 0:6].sum(axis=1)
```

```python
fig, ax = plt.subplots(figsize=(14, 7))

bar_width = 20
bottom = None

for col in monthly_sum.columns[:6]:
    ax.bar(monthly_sum.index, monthly_sum[col], label=col, bottom=bottom, width=bar_width)
    if bottom is None:
        bottom = monthly_sum[col]
    else:
        bottom += monthly_sum[col]

ax.plot(monthly_sum.index, monthly_sum['Total'], color='black', marker='o', label='总销售量')
ax.set_xlabel('日期')
ax.set_ylabel('销售量')
ax.legend()
plt.xticks(rotation=45)

plt.title('各类蔬菜月销售量变化图（2020-2023）')
plt.grid(True)
plt.tight_layout()
plt.savefig(f'各类蔬菜月销售量变化图（2020-2023）.png')
plt.show()
```

```python
for col in df.columns:
    decomposition = seasonal_decompose(df[col], model='additive', period=31)
    seasonal = decomposition.seasonal
    decomposition.plot()
    plt.savefig(f'{col}销量数据季节性分解.png')
```
