```python
import numpy as np
import pandas as pd
df=pd.read_excel("附件/蔬菜单品日销量.xlsx", index_col=0).T
df2=pd.read_excel("2.xlsx", index_col=0)
```

```python
df1=pd.DataFrame()
df1['总销量'] = df.iloc[:, :].sum(axis=1)
df1[['打折次数','销售次数', '平均定价', '定价汇总']]=0
```

```python
# 注意：由于数据有246*877000，这个代码将近需要1h，可以用i来判断进度
i=0
for product_ID in df2['product_code'].unique():
    print(i)
    i+=1
    for index, row in df2.iterrows():
        if row['product_code'] == product_ID:
            if row['is_discount'] == 0:
                df1.loc[product_ID, '打折次数']+=1
            df1.loc[product_ID, '销售次数']+=1
            df1.loc[product_ID, '定价汇总']+=row['unit_price']
        
```

```python
df1.to_excel('所有单品信息（总销量-打折次数—日定价-销售次数）.xlsx', index=True)
```

```python
# 统计合并该段时间内附件四出现的单品的损耗率
import pandas as pd

df1=pd.read_excel("所有单品信息（总销量-打折次数—日定价-销售次数）.xlsx", index_col=0) # 该文件未被包含在文件夹内，有完整版数据
df3=pd.read_excel("4_1.xlsx")
# 不需要记录单品的名称
del df3['product_name']
# 将已经统计的表添加一列‘product_code’便于合并
df1['product_code']=df1.index
df1['平均定价']=df1['定价汇总']/df1['销售次数']
merged_df = df1.merge(df3, on=['product_code', 'product_code'], how='inner')
merged_df.index=merged_df['product_code']
del merged_df['product_code']
del merged_df['定价汇总']
del df1['product_code']
merged_df.rename(columns={'loss_rate': '损耗率'}, inplace=True)
merged_df.to_excel('完整-所有单品信息（总销量-打折次数—日定价-销售次数）.xlsx', index=True)
```
